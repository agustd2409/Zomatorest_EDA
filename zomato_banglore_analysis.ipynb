```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['url','phone','address','menu_item','dish_liked','reviews_list'],axis = 1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.shape
```

```python
df.info()
```

Cleaning "Rate" column

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate'].isnull().sum()
```

Replacing null values in rate column with mean

```python
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

Dropping null values

```python
df.dropna(inplace=True)
df.head()
```

```python
df.isnull().sum()
```

Renaming columns

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
#location and listed_in(city) both are locations, so keep one

df = df.drop(['listed_in(city)'], axis=1)
```

```python
df.head()
```

```python
df['Cost2plates'].unique()
```

Removing ',' from Cost2plates

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',' , '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

Cleaning "rest_type" column

```python
df['rest_type'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

Making rest_types_lessthan1000 in frequency as 'others'

```python
def handle_rest(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_rest)
df['rest_type'].head()
        
```

```python
df['rest_type'].value_counts()
```

Cleaning location column

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = False)
```

```python
location_lessthan300 = location[location<300]
```

```python
def handle_loc(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
    
df['location'] = df['location'].apply(handle_loc)
df['location'].head()
```

```python
df['location'].value_counts()
```

Cleaning cuisines column

```python
cuisines = df['cuisines'].value_counts(ascending = False)
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].head()
```

```python
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df['Type'].value_counts()
```
